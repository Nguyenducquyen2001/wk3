```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

food = ctrl.Antecedent(np.arange(0,11,1),'Food')
service = ctrl.Antecedent(np.arange(0,11,1),'service')
tip =ctrl.Consequent(np.arange(10,31,1),'Tip')
food['poor']=fuzz.trimf(food.universe,[0,0,5])
food['avenrage']=fuzz.trimf(food.universe,[0,5,10])
food['good']=fuzz.trimf(food.universe,[5,10,10])

service['poor']=fuzz.trimf(service.universe,[0,0,5])
service['avenrage']=fuzz.trimf(service.universe,[0,5,10])
service['good']=fuzz.trimf(service.universe,[5,10,10])

tip['less']=fuzz.trimf(tip.universe,[10,10,20])
tip['avenrage']=fuzz.trimf(tip.universe,[10,20,30])
tip['much']=fuzz.trimf(tip.universe,[20,30,30])
rule = ctrl.Rule(food['good'] & service['poor'],tip['less'])
rule1 = ctrl.Rule(food['good'] & service['avenrage'],tip['avenrage'])
rule2 = ctrl.Rule(food['good'] & service['good'],tip['much'])

rule3 = ctrl.Rule(food['avenrage'] & service['poor'],tip['less'])
rule4 = ctrl.Rule(food['avenrage'] & service['avenrage'],tip['avenrage'])
rule5 = ctrl.Rule(food['avenrage'] & service['good'],tip['much'])

rule6 = ctrl.Rule(food['poor'] & service['poor'],tip['less'])
rule7 = ctrl.Rule(food['poor'] & service['avenrage'],tip['avenrage'])
rule8 = ctrl.Rule(food['poor'] & service['good'],tip['avenrage'])

tipping_ctrl=ctrl.ControlSystem([rule,rule1,rule2,rule3,rule4,rule,rule6,rule7,rule8])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['Food']= 7.8
tipping.input['service']= 9.8
tipping.compute()
print(tipping.output['Tip'])

food.view()
service.view()
tip.view()
```

```python
#điều khiển bơm nước

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

C = ctrl.Antecedent(np.arange(0,2001,1),'C')
T = ctrl.Antecedent(np.arange(0,2001,1),'T')
Tg =ctrl.Consequent(np.arange(0,31,1),'Tg')
C['Ce']=fuzz.trimf(C.universe,[0,0,1000])
C['Cm']=fuzz.trimf(C.universe,[0,1000,2000])
C['Cf']=fuzz.trimf(C.universe,[1000,2000,2000])

T['Te']=fuzz.trimf(T.universe,[0,0,1000])
T['Tf']=fuzz.trimf(T.universe,[1000,2000,2000])

Tg['L']=fuzz.trapmf(Tg.universe,[0,0,5,10])
Tg['M']=fuzz.trapmf(Tg.universe,[5,10,20,25])
Tg['S']=fuzz.trapmf(Tg.universe,[20,25,30,30])

C.view()
T.view()
Tg.view()

rule0 = ctrl.Rule(C['Ce'] & T['Te'],Tg['L'])
rule1 = ctrl.Rule(C['Ce'] & T['Tf'],Tg['M'])
rule2 = ctrl.Rule(C['Cm'] & T['Te'],Tg['L'])
rule3 = ctrl.Rule(C['Cm'] & T['Tf'],Tg['S'])
rule4 = ctrl.Rule(C['Cf'] & T['Te'],Tg['M'])
rule5 = ctrl.Rule(C['Cf'] & T['Tf'],Tg['S'])

timing_ctrl=ctrl.ControlSystem([rule0,rule1,rule2,rule3,rule4])
timing = ctrl.ControlSystemSimulation(timing_ctrl)

nhapnuoc= input('nhap luong nuoc co trong bon (ml): ')
timing.input['C']= int(nhapnuoc)
timing.input['T']= int(nhapnuoc)
timing.compute()
print(timing.output['Tg'])
```

```python
# bài tập nồi cơm điện 

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

time = ctrl.Antecedent(np.arange(0,181,1),'time')
rice = ctrl.Antecedent(np.arange(200,1001,1),'rice')
power =ctrl.Consequent(np.arange(0,101,1),'power')

time['vf']=fuzz.trimf(time.universe,[0,15,40])
time['f']=fuzz.trimf(time.universe,[15,40,90])
time['av']=fuzz.trimf(time.universe,[40,90,140])
time['s']=fuzz.trimf(time.universe,[90,140,165])
time['vl']=fuzz.trimf(time.universe,[140,165,180])

rice['vl']=fuzz.trimf(rice.universe,[200,300,400])
rice['l']=fuzz.trimf(rice.universe,[300,400,600])
rice['av']=fuzz.trimf(rice.universe,[400,600,800])
rice['m']=fuzz.trimf(rice.universe,[600,800,900])
rice['vm']=fuzz.trimf(rice.universe,[800,900,1000])

power['v']=fuzz.trimf(power.universe,[0,10,25])
power['s']=fuzz.trimf(power.universe,[10,25,50])
power['av']=fuzz.trimf(power.universe,[25,50,75])
power['h']=fuzz.trimf(power.universe,[50,75,90])
power['vh']=fuzz.trimf(power.universe,[75,90,100])

time.view()
rice.view()
power.view()

rule0 = ctrl.Rule(time['vf'] & rice['vl'], power['v'])
rule1 = ctrl.Rule(time['vf'] & rice['l'], power['v'])
rule2 = ctrl.Rule(time['vf'] & rice['av'], power['s'])
rule3 = ctrl.Rule(time['vf'] & rice['m'], power['s'])
rule4 = ctrl.Rule(time['vf'] & rice['vm'], power['av'])

rule5 = ctrl.Rule(time['f'] & rice['vl'], power['v'])
rule6 = ctrl.Rule(time['f'] & rice['l'], power['s'])
rule7 = ctrl.Rule(time['f'] & rice['av'], power['s'])
rule8 = ctrl.Rule(time['f'] & rice['m'], power['av'])
rule9 = ctrl.Rule(time['f'] & rice['vm'], power['av'])

rule10 = ctrl.Rule(time['av'] & rice['vl'], power['s'])
rule11 = ctrl.Rule(time['av'] & rice['l'], power['s'])
rule12 = ctrl.Rule(time['av'] & rice['av'], power['av'])
rule13 = ctrl.Rule(time['av'] & rice['m'], power['av'])
rule14 = ctrl.Rule(time['av'] & rice['vm'], power['h'])

rule15 = ctrl.Rule(time['s'] & rice['vl'], power['s'])
rule16 = ctrl.Rule(time['s'] & rice['l'], power['av'])
rule17 = ctrl.Rule(time['s'] & rice['av'], power['av'])
rule18 = ctrl.Rule(time['s'] & rice['m'], power['h'])
rule19 = ctrl.Rule(time['s'] & rice['vm'], power['h'])

rule20 = ctrl.Rule(time['vl'] & rice['vl'], power['av'])
rule21 = ctrl.Rule(time['vl'] & rice['l'], power['av'])
rule22 = ctrl.Rule(time['vl'] & rice['av'], power['h'])
rule23 = ctrl.Rule(time['vl'] & rice['m'], power['h'])
rule24 = ctrl.Rule(time['vl'] & rice['vm'], power['vh'])

power_ctrl=ctrl.ControlSystem([rule0,rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24])
power_out = ctrl.ControlSystemSimulation(power_ctrl)

nhaptime= input('nhap thoi gian nau: (phut)')
nhaprice = input('nhap luong gao: (ml)')
power_out.input['time']= int(nhaptime)
power_out.input['rice']= int(nhaprice)
power_out.compute()
print(power_out.output['power'])
```
